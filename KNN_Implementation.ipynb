# Train Test Split

```python
x= df.drop('Outcome',axis=1)
y= df['Outcome']
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=24,stratify=y)
```

```python
x_test
```

```python
x_train
```

```python
Model_Training:
```

```python
knn_clf=KNeighborsClassifier(n_neighbors=5,p=2)
knn_clf.fit(x_train,y_train)
```

```python
#Evaluation on Testing data:

y_pred=knn_clf.predict(x_test)

cnf_matrix=confusion_matrix(y_test,y_pred)
print('Confusion Matric is:\n',cnf_matrix)

acc_score=accuracy_score(y_test,y_pred)
print("Accuracy:\n",acc_score)

clf_report= classification_report(y_test,y_pred)
print("Classification Report :\n",clf_report)
```

```python
#Evaluation on Training data:

y_train_pred=knn_clf.predict(x_train)

cnf_matrix=confusion_matrix(y_train,y_train_pred)
print('Confusion Matric is:\n',cnf_matrix)

acc_score=accuracy_score(y_train,y_train_pred)
print("Accuracy:\n",acc_score)

clf_report= classification_report(y_train,y_train_pred)
print("Classification Report :\n",clf_report)
```

```python
Hyperparameter Tunning:
```

```python
knn_clf=KNeighborsClassifier(p=2,n_neighbors=5)

hyperparameters= {'n_neighbors':np.arange(2,30),'p':[1,2]}

gscv_knn_clf= GridSearchCV(knn_clf,hyperparameters,cv=5)
gscv_knn_clf.fit(x_train,y_train)
```

```python
gscv_knn_clf.best_estimator_
```

```python
gscv_knn_clf.best_params_
```

```python
knn_clf=gscv_knn_clf.best_estimator_
knn_clf.fit(x_train,y_train)
```

```python
#Evaluation on Testing data:

y_pred=knn_clf.predict(x_test)

cnf_matrix=confusion_matrix(y_test,y_pred)
print('Confusion Matric is:\n',cnf_matrix)

acc_score=accuracy_score(y_test,y_pred)
print("Accuracy:\n",acc_score)

clf_report= classification_report(y_test,y_pred)
print("Classification Report :\n",clf_report)
```

```python
#Evaluation on Training data:

y_train_pred=knn_clf.predict(x_train)

cnf_matrix=confusion_matrix(y_train,y_train_pred)
print('Confusion Matric is:\n',cnf_matrix)

acc_score=accuracy_score(y_train,y_train_pred)
print("Accuracy:\n",acc_score)

clf_report= classification_report(y_train,y_train_pred)
print("Classification Report :\n",clf_report)
```
